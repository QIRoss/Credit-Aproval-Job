### Imports

```python
import pandas as pd
import sklearn.linear_model as lm
```

### Leitura de arquivos

```python
df_from_eda = pd.read_csv('data/desafio 2/hp_from_eda.csv')

df_from_eda
```

### Split

```python
train = df_from_eda[df_from_eda['origem'] == 'treino']
test = df_from_eda[df_from_eda['origem'] == 'teste']

# dropar variÃ¡vel alvo do teste e origem

test.drop(columns=['preco', 'origem'], inplace=True)
train.drop(columns='origem', inplace=True)
```

```python
test
```

```python
train
```

```python
y_train = train['preco'].to_numpy()
print(set(y_train))
X_train = train.drop(columns=['preco']).select_dtypes(include='number').to_numpy()
X_test = test.select_dtypes(include='number').to_numpy()
print(f'X_test len: {len(X_test)}')


print(X_train.shape)
print(y_train.shape)
```

### Model

#### Ridge regression

```python
reg = lm.Ridge(alpha=20.0)
reg.fit(X_train, y_train)
pred = reg.predict(X_test)

pred
```

```python
prediction_df = pd.DataFrame({'Id' : test['Id'].to_numpy(),
                              'preco' : pred})
```

```python
prediction_df
```

```python
prediction_df['preco'] = prediction_df['preco'].round(2)
```

```python
prediction_df
```

```python
prediction_df.to_csv('data/desafio 2/predictions/ridge_5.csv', index=False)
```

### Lasso

```python
reg_lass = lm.Lasso(alpha=0.5)
reg_lass.fit(X_train, y_train)
pred = reg_lass.predict(X_test)
```

```python
predict_df = pd.DataFrame({'Id' : test['Id'].to_numpy(),
                              'preco' : pred})
```

```python
predict_df
```

```python
predict_df['preco'] = predict_df['preco'].round(2)
```

```python
predict_df.to_csv('data/desafio 2/predictions/lasso_1.csv', index=False)
```

### LinearRegression

```python
lr = lm.LinearRegression()
lr.fit(X_train, y_train)
pred = lr.predict(X_test)
```

```python
predict_df = pd.DataFrame({'Id' : test['Id'].to_numpy(),
                              'preco' : pred})
```

```python
predict_df['preco'] = predict_df['preco'].round(2)
```

```python
predict_df.to_csv('data/desafio 2/predictions/lr_1.csv', index=False)
```
